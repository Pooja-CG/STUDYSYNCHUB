```python
import imaplib
import email
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
import time

def send_autoresponse(sender_email):
    # Email content for autoresponse
    subject = "Autoresponse"
    body = "Thank you for your email. We will get back to you shortly."

    # Create message
    message = MIMEMultipart()
    message["From"] = "studysynchub24@gmail.com"
    message["To"] = sender_email
    message["Subject"] = subject
    message.attach(MIMEText(body, "plain"))

    # Connect to SMTP server and send autoresponse
    with smtplib.SMTP("smtp.gmail.com", 587) as server:
        server.starttls()
        server.login("studysynchub24@gmail.com", "wblu hgsb pwrm ofek")
        server.sendmail("studysynchub24@gmail.com", sender_email, message.as_string())

def check_emails():
    # Connect to IMAP server
    mail = imaplib.IMAP4_SSL("imap.gmail.com")
    mail.login("studysynchub24@gmail.com", "wblu hgsb pwrm ofek")
    mail.select("inbox")

    # Search for unseen emails
    result, data = mail.search(None, "UNSEEN")
    if result == "OK":
        for num in data[0].split():
            # Fetch email data
            result, data = mail.fetch(num, "(RFC822)")
            if result == "OK":
                raw_email = data[0][1]
                msg = email.message_from_bytes(raw_email)
                sender_email = msg["From"]

                # Send autoresponse
                send_autoresponse(sender_email)

    mail.close()
    mail.logout()


```

```python
# Check emails every minute and send autoresponse if new emails are found
while True:
    check_emails()
    time.sleep(0.1)
```

```python
import re

def is_valid_gmail(email):
    # Define the regular expression pattern for a Gmail address
    pattern = r'^[a-zA-Z0-9._%+-]+@gmail\.com$'

    # Use the re.match() function to check if the email matches the pattern
    match = re.match(pattern, email)

    # If there is a match, the email is valid; otherwise, it's not valid
    return bool(match)

# Example usage:
email_to_check = 'sherlys@gmail.com'
if is_valid_gmail(email_to_check):
    print(f"{email_to_check} is a valid Gmail address.")
else:
    print(f"{email_to_check} is not a valid Gmail address.")
```
